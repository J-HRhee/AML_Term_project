```python
import pandas as pd
import numpy as np

data = pd.read_csv("datasets/WA_Fn-UseC_-Telco-Customer-Churn.csv")
data
```

null값 찾기

```python
data.info()
```

```python
data.isnull().sum()
```

공백 값이 있는 열 제외

```python
data = data[data['TotalCharges']!=' '] 
data # 11개 제거됨
```

```python
#ID는 예측할 때는 필요가 없으니 빼서 따로 저장.
ID = data['customerID'].copy()
data = data.drop(["customerID"], axis=1)
```

```python
data.info()
```

TotalCharges를 숫자로 변환

```python
data['TotalCharges'] = pd.to_numeric(data["TotalCharges"]) # 숫자로 변환된 것을 확인
data.info()
```

데이터 준비과정

```python
# 숫자형 데이터
data_num = data[["tenure","MonthlyCharges","TotalCharges"]]

# 문자형 데이터
data_str = data.drop(["tenure","MonthlyCharges","TotalCharges"], axis=1)

# 문자형-이진(senior citizen은 숫자형이긴 해도 값이 0/1밖에 없기 때문에 이곳에 포함시켜도 무방하다)
data_ordinal = data_str[["SeniorCitizen","gender","Partner", "Dependents", "PhoneService", "Churn"]] # Y/N  or F/M

# 문자형-순서가 없는 데이터
data_hot = data_str.drop(["SeniorCitizen","gender","Partner", "Dependents", "PhoneService", "Churn"], axis=1) # Y/N/No service
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder

num_attribs = list(data_num)
ordinal_attribs = list(data_ordinal)
hot_attribs = list(data_hot)

Pipeline_1 = ColumnTransformer([
        ("ordinal", OrdinalEncoder(), ordinal_attribs), # 남자는1, 여자는0으로 맵핑됨.
        ("hot", OneHotEncoder(sparse=False), hot_attribs),
        ("num", StandardScaler(), num_attribs), # 숫자 데이터 스케일링(standard)    
    ])

Pipeline_2 = ColumnTransformer([
        ("ordinal", OrdinalEncoder(), ordinal_attribs), # 남자는1, 여자는0으로 맵핑됨.
        ("hot", OrdinalEncoder(), hot_attribs),  
        ("num", RobustScaler(), num_attribs), # 숫자 데이터 스케일링(Robust)
    ])

data_prepared_1 = pd.DataFrame(Pipeline_1.fit_transform(data)) 
data_prepared_2 = pd.DataFrame(Pipeline_2.fit_transform(data)) 
data_prepared_2
```

훈련셋, 검증셋 생성

```python
from sklearn.model_selection import train_test_split
#train_set, test_set = train_test_split(data_prepared_1, test_size=0.2, random_state=42)
train_set, test_set = train_test_split(data_prepared_2, test_size=0.2, random_state=42)

# data_prepared의 5번 컬럼이 data['Churn']이다.
train_x = train_set.drop([5], axis=1) 
train_y = train_set[5] # 훈련 셋 레이블

test_x = test_set.drop([5], axis=1)
test_y = test_set[5] # 검증 셋 레이블
```

RandomForestClassifier

```python
comparison=[]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

RFC = RandomForestClassifier(random_state=42, max_depth= 4)
RFC.fit(train_x, train_y) # 훈련 셋에 적용
RFC_predict = RFC.predict(train_x) 
RFC_predict_test = RFC.predict(test_x)

AccuracyScore = []
AccuracyScore.append(accuracy_score(RFC_predict, train_y))
AccuracyScore.append(accuracy_score(RFC_predict_test, test_y))

comparison.append(AccuracyScore)

print('train_set: ', AccuracyScore[0]) # 훈련 셋 예측 결과
print('test_set: ', AccuracyScore[1]) # 검증 셋 적용 결과
```

BaggingClassifier

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

bagC = BaggingClassifier(
    DecisionTreeClassifier(random_state=42),random_state=42, n_estimators=500, max_samples=100)
bagC.fit(train_x, train_y)

bagC_predict = bagC.predict(train_x)
bagC_predict_test = bagC.predict(test_x)

AccuracyScore = []
AccuracyScore.append(accuracy_score(bagC_predict, train_y))
AccuracyScore.append(accuracy_score(bagC_predict_test, test_y))

comparison.append(AccuracyScore)

print('train_set: ', AccuracyScore[0]) # 훈련 셋 예측 결과
print('test_set: ', AccuracyScore[1]) # 검증 셋 적용 결과
```

부스팅: XGBoost(조기종료)

```python
import xgboost as xgb

XGB = xgb.XGBRegressor(random_state=42)
XGB.fit(train_x, train_y,
        eval_set=[(test_x, test_y)], early_stopping_rounds=10)
XGB_predict = XGB.predict(train_x)
XGB_predict_test = XGB.predict(test_x)

AccuracyScore = []
AccuracyScore.append(accuracy_score([ 1 if x > 0.5 else 0 for x in XGB_predict], train_y))
AccuracyScore.append(accuracy_score([ 1 if x > 0.5 else 0 for x in XGB_predict_test], test_y))

comparison.append(AccuracyScore)

print('train_set: ', AccuracyScore[0]) # 훈련 셋 예측 결과
print('test_set: ', AccuracyScore[1]) # 검증 셋 적용 결과
```

부스팅: ADABoost(얕은 분류기 200회)

```python
from sklearn.ensemble import AdaBoostClassifier

adaC = AdaBoostClassifier(
    DecisionTreeClassifier(max_depth=1), n_estimators=200)
adaC.fit(train_x, train_y)
adaC_predict = adaC.predict(train_x)
adaC_predict_test = adaC.predict(test_x)

AccuracyScore = []
AccuracyScore.append(accuracy_score(adaC_predict, train_y))
AccuracyScore.append(accuracy_score(adaC_predict_test, test_y))

comparison.append(AccuracyScore)

print('train_set: ', AccuracyScore[0]) # 훈련 셋 예측 결과
print('test_set: ', AccuracyScore[1]) # 검증 셋 적용 결과
```

LinearRegression

```python
# 오류가 날 때 이 부분을 다시 실행하면 된다.

from sklearn.linear_model import LinearRegression

LinR = LinearRegression()
LinR.fit(train_x, train_y)
LinR_predict = LinR.predict(train_x)
LinR_predict_test = LinR.predict(test_x)

AccuracyScore = []
AccuracyScore.append(accuracy_score([ 1 if x > 0.5 else 0 for x in LinR_predict], train_y))
AccuracyScore.append(accuracy_score([ 1 if x > 0.5 else 0 for x in LinR_predict_test], test_y))

comparison.append(AccuracyScore)

print('train_set: ', AccuracyScore[0]) # 훈련 셋 예측 결과
print('test_set: ', AccuracyScore[1]) # 검증 셋 적용 결과
```

다항회귀

```python
from sklearn.preprocessing import PolynomialFeatures

poly_features = PolynomialFeatures(degree=2, include_bias=False)
train_x_poly = poly_features.fit_transform(train_x)
test_x_poly = poly_features.fit_transform(test_x)

LinR.fit(train_x_poly, train_y) 
LinR_poly_predict = LinR.predict(train_x_poly)
LinR_poly_predict_test = LinR.predict(test_x_poly)

AccuracyScore = []
AccuracyScore.append(accuracy_score([ 1 if x > 0.5 else 0 for x in LinR_poly_predict], train_y))
AccuracyScore.append(accuracy_score([ 1 if x > 0.5 else 0 for x in LinR_poly_predict_test], test_y))

comparison.append(AccuracyScore)

print('train_set: ', AccuracyScore[0]) # 훈련 셋 예측 결과
print('test_set: ', AccuracyScore[1]) # 검증 셋 적용 결과
```

```python
from sklearn.svm import SVC

svc = SVC(gamma="scale", random_state=42)
svc.fit(train_x, train_y)
svc_predict = svc.predict(train_x)
svc_predict_test = svc.predict(test_x)

AccuracyScore = []
AccuracyScore.append(accuracy_score(svc_predict, train_y))
AccuracyScore.append(accuracy_score(svc_predict_test, test_y))

comparison.append(AccuracyScore)

print('train_set: ', AccuracyScore[0]) # 훈련 셋 예측 결과
print('test_set: ', AccuracyScore[1]) # 검증 셋 적용 결과
```
